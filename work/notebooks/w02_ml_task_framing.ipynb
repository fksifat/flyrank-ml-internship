## 1. My lane as an ML task (type)

For the **Refresh / Content Opportunity Scoring** lane, the task type is **Ranking / Scoring** (though built on top of a classification probability). The team has limited capacity, so they don't just need a list of 'declining' pages—they need an ordered queue. Ranking surfaces the highest-priority pages to the top, ensuring the editor's time is spent on the most promising opportunities first.

```python
# No code strictly necessary here, but we can verify task type properties.
print("Task type chosen: Ranking / Scoring based on probability outputs.")
```

## 2. Target or proxy

Our proxy target is **whether the page is in decline**. In the starter dataset, this is represented by the `trend_direction` being 'down'. This comes from an **observed outcome** (historical changes in traffic metrics), not from a human-defined decision rule like a `priority_score`.

```python
import pandas as pd
df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")
df["is_declining"] = (df["trend_direction"] == "down").astype(int)
print("Target Proxy:")
display(df[["content_id", "trend_direction", "is_declining"]].head(3))
```

## 3. Success metric

Our success metric is **Precision@K (e.g., Precision@50)**. Because the reviewer can only check a limited number of pages (say, 50), we only care about the accuracy at the very top of our ranked list. A 'good' number means a high percentage of the top 50 recommendations are genuinely declining pages that warrant attention.

```python
# Example of what the metric needs:
k = 50
print(f"Metric: Precision@{k}")
print(f"Goal: Maximize the fraction of true positives in the top {k} recommended pages.")
```

## 4. The unit of analysis, as a real dataframe

The unit of analysis is **one row = one content item (page) for a specific client**. Each row aggregates the performance over a 90-day window.

```python
print(f"Dataframe shape: {df.shape[0]} rows, {df.shape[1]} columns")
print("Unit of analysis (one row = one content item):\n")
display(df[["content_id", "client_id", "content_type", "impressions_90d", "is_declining"]].head(5))
```

## 5. Why ML beats a fixed rule here

A fixed rule (e.g., `impressions > 1000` AND `content_age > 180`) is rigid and creates hard cliffs. It treats all pages exactly the same regardless of nuance. Machine Learning (like a Random Forest) can learn non-linear interactions—for example, that low CTR is acceptable for informational content on page 2, but critical for transactional content on page 1. It weighs dozens of weak signals into one strong, nuanced priority score.
