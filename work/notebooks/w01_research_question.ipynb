## 3. Quick look at the data (2-3 real numbers)

Let's see how many pages we are dealing with. If the number of declining or stale pages is small, we could just review them all manually. If it's large, we desperately need ranking.

```python
import pandas as pd
# Note: working from work/notebooks, the data is two levels up
df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")

total_pages = len(df)
declining_pages = (df["trend_direction"] == "down").sum()
stale_pages = (df["content_age_days"] >= 180).sum()

print(f"Total pages in sample: {total_pages:,}")
print(f"Pages showing a downward trend: {declining_pages:,} ({declining_pages/total_pages:.1%})")
print(f"Pages older than 180 days: {stale_pages:,} ({stale_pages/total_pages:.1%})")
print("\nWith ~11k declining pages and ~16k stale pages, a human reviewer cannot check them all. We need a ranking model to surface the top 50.")
```
